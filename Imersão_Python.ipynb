```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/base de acoes_pura.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/base de acoes_pura.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/base de acoes_pura.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/base de acoes_pura.xlsx", sheet_name="Chatgpt 2")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final','Var. Dia':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['Valor_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['Valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
# Calculando o maior valor
maior = df_principal['Variacao_rs'].max()

# Calculando o menor valor
menor = df_principal['Variacao_rs'].min()

# Calculando a média
media = df_principal['Variacao_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais Por resultado')
fig.update_traces(hovertemplate='Resultado: %{x}<br>Variacao: R$ %{text:,.2f}')
fig.update_traces(texttemplate='R$ %{text:,.2f}', textposition='inside')
fig.show()
```

```python
df_analise_segmento_subiu = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
fig = px.pie(df_analise_segmento_subiu, values='Variacao_rs', names='Segmento', title='Variação Reais Por Segmento')
fig.update_traces(hovertemplate='Segmento: %{label}<br>Variacao: R$ %{value:,.2f}')
fig.update_traces(textinfo='label+percent', textposition='inside')
fig.show()
```

```python
df_analise_idade = df_principal.groupby('Cat_idade')['Variacao_rs'].sum().reset_index()
df_analise_idade['Análise por faixa etária'] = df_analise_idade['Cat_idade'].apply(lambda x: 'Empresas com mais de 100 anos' if x == 'Mais de 100' else ('Empresas com menos de 50 anos' if x == 'Menos de 50' else 'Empresas entre 50 e 100 anos'))
df_analise_idade['Quantidade de empresas'] = df_principal.groupby('Cat_idade')['Ativo'].count().reset_index()['Ativo']
df_analise_idade['variação em reais'] = 'R$ ' + df_analise_idade['Variacao_rs'].apply(lambda x: '{:,.2f}'.format(x).replace(',', '.'))
df_analise_idade = df_analise_idade[['Análise por faixa etária', 'variação em reais', 'Quantidade de empresas']]
print(df_analise_idade.to_string())
```

```python
fig = px.bar(df_analise_idade, x='Análise por faixa etária', y='variação em reais', text='Quantidade de empresas', title='Variação Reais Por Faixa Etária')
fig.update_traces(hovertemplate='Faixa Etária: %{x}<br>Variação: %{y}<br>Quantidade de empresas: %{text}')
fig.update_traces(texttemplate='%{text}', textposition='outside')
fig.show()
```
